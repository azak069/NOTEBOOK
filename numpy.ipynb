# Properties
* **ndarray.ndim**: dimensi (rank) dari sebuah numpy array.
* **ndarray.shape**: jumlah baris dan kolum dari sebuah numpy array.
* **ndarray.size**: jumlah elemen di dalam numpy array.
* **ndarray.dtype**: tipe data dari elemen yang ada di numpy array.
* **ndarray.itemsize**: ukuran elemen (bytes) yang ada di numpy array.
* **ndarray.data**: buffer untuk menyimpan elemen-elemen di dalam numpy array.

```python
a = np.array([1, 2, 3])
b = np.array([(1, 2, 3), (4, 5, 6)])
```

```python
print(a.ndim)
print(a.shape)
print(a.size)
print(a.dtype)
print(a.itemsize)
```

```python
print(b.ndim)
print(b.shape)
print(b.size)
print(b.dtype)
print(b.itemsize)
```

# Create Numpy Arrays

* **np.zeros**: isinya semua 0, berikan baris dan kolom
* **np.ones**: isinya semua 1, berikan baris dan kolom
* **np.arange**: isi scope (dari berapa ke berapa) dan bisa selisih antar angka
* **np.linspace**: isi scope dan jumlah elemen yang diinginkan
* **np.random.random**: isi angka acak, berikan baris dan kolom

```python
np.zeros((3, 3)) # membuat array 3x3 berisi nol
```

```python
np.ones((3, 3)) # membuat array 3x3 berisi satu
```

```python
np.arange(1, 2, 0.2) # membuat array dengan nilai dari 1 sampai 2 dengan step 0.2
```

```python
np.linspace(1, 3, 10) # membuat array dengan 10 nilai yang terdistribusi merata antara 1 dan 3
```

```python
np.random.random((2, 3)) # membuat array 2x3 dengan nilai random antara 0 dan 1
```

# Shape and Operation

* **reshape**: mengubah dimensi/shape dari sebuah numpy array.
* **vstack**: stack (menggabungkan) numpy arrays secara vertikal.
* **hstack**: stack numpy arrays secara horizontal.
* **vsplit**: split (membagi) numpy arrays secara vertikal.
* **hsplit**: split numpy arrays secara horizontal.

```python
a = np.array([(1, 2, 3), (4, 5, 6)])
a
```

```python
a = a.reshape(3, -1) # mengubah ke 3 baris dan kolomnya otomatis disesuaikan
a
```

```python
b = np.array([(7, 8), (9, 10)])
b
```

```python
np.vstack((a, b)) # menggabungkan array a dan b secara vertikal
```

```python
c = np.array([(11, 12, 13), (14, 15, 16)])
c
```

```python
a = a.reshape(2, -1) # mengubah ke 2 baris dan kolomnya otomatis disesuaikan
a
```

```python
d = np.hstack((a, c)) # menggabungkan array a dan c secara horizontal
d
```

```python
d
```

```python
np.vsplit(d, 2) # membagi array d menjadi 2 bagian secara vertikal
```

```python
np.hsplit(d, 3) # membagi array d menjadi 3 bagian secara horizontal
```

```python
np.hsplit(d, (1, 3)) # split/bagi setelah kolom 1 dan kolom 3
```

# Indexing

```python
data = np.arange(100, 200) # membuat array dengan nilai 100-199
data
```

```python
data[0]
```

```python
fives = np.arange(0, 100, 5) # membuat array dengan kelipatan 5 dari 0-95
fives
```

```python
data[fives] # ambil elemen di index sesuai nilai pada array fives
```

```python
data2D = data.reshape(10, -1) # mengubah data menjadi array 2D dengan 10 baris
data2D
```

```python
data2D[2] # ambil baris ke-2
```

```python
data2D[2, 1] # ambil elemen di baris ke-2 dan kolom ke-1
```

```python
data2D[-2, -1] # ambil elemen di baris ke-2 dari akhir dan kolom terakhir
```

```python
data2D[:, 3] # ambil semua baris di kolom ke-3
```

```python
# baris 1, 2 dan 3 
# kolom 3 dan 4
data2D[1:4, 3:5] # ambil sub-array dari baris 1-3 dan kolom 3-4
```

# Mathematics

```python
# operasi matematika dasar pada array numpy
numbers = np.random.random((5, 5)) # membuat array 5x5 dengan nilai random antara 0 dan 1
numbers
```

```python
print(np.amin(numbers)) # nilai minimum dari array
print(np.amax(numbers)) # nilai maksimum dari array
print(np.average(numbers)) # nilai rata-rata dari array
print(np.mean(numbers)) # nilai mean dari array
print(np.median(numbers)) # nilai median dari array
print(np.sum(numbers)) # jumlah semua elemen dalam array
```

```python
np.sqrt(numbers) # menghitung akar kuadrat dari setiap elemen dalam array
```

```python
np.sin(numbers) # menghitung nilai sinus dari setiap elemen dalam array
```

```python
np.cos(numbers) # menghitung nilai cosinus dari setiap elemen dalam array
```

```python
np.tan(numbers) # menghitung nilai tangen dari setiap elemen dalam array
```

```python
sample = np.arange(1, 21) # membuat array dengan nilai 1-20
sample
```

```python
sample2 = np.arange(51, 71) # membuat array dengan nilai 51-70
sample2
```

```python
sample + sample2 # penjumlahan elemen array sesuai index
```

```python
sample2 - sample # pengurangan elemen array sesuai index
```

```python
sample * sample2 # perkalian elemen array sesuai index
```

```python
sample / sample2 # pembagian elemen array sesuai index
```

# Matrix

```python
# operasi matriks dengan numpy
matrix = np.random.randint(1, 50, 20) # membuat array dengan 20 bilangan bulat random antara 1-50 (random integer)
matrix = matrix.reshape(4, -1) # mengubah array menjadi 4x5 berbentuk 2 Dimensi
matrix
```

```python
matrix.transpose() # mentranspose/memutar matriks (baris jadi kolom dan kolom jadi baris)
```

```python
np.dot(matrix, -1) # mengalikan matriks dengan skalar -1 (setiap elemen dikalikan -1)
```
